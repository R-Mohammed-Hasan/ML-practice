```python
import asyncio

from browser_use import Agent, Browser, BrowserConfig
from dotenv import load_dotenv
from transformers import pipeline  # Importing transformers pipeline

load_dotenv()

browser = Browser(
    config=BrowserConfig(
        # Specify the path to your Chrome executable
        chrome_instance_path="/Applications/Google Chrome.app/Contents/MacOS/Google Chrome",  # macOS path
        # For Windows, typically: 'C:\\Program Files\\Google\\Chrome\\Application\\chrome.exe'
        # For Linux, typically: '/usr/bin/google-chrome'
    )
)

# llm = ChatOpenAI(model="gpt-4o")
llm = pipeline(
    "text-generation",
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
)


async def main():
    """
    The function `main` creates an `Agent` object to compare the price of two products and then runs the
    agent to get the result.
    """
    """
    The main function creates an Agent object to compare the price of GPT-4o and DeepSeek-V3, then runs
    the agent and prints the result.
    """
    agent = Agent(
        task="Compare the price of gpt-4o and DeepSeek-V3",
        llm=llm,
        # browser=browser,
    )
    result = await agent.run()
    print(result)


asyncio.run(main())
```
